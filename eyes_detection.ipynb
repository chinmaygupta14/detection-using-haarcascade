```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread(r'C:\Users\Acer\Desktop\camera\Camera\faces.jpg')
fix= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix)
```

```python
eye_classifier=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eyes(fix_img):
    eyes_rects=eye_classifier.detectMultiScale(fix_img,1.3,5)
    
    for(x,y,w,h) in eyes_rects:
        cv2.rectangle(fix_img,(x,y),(x+w,y+h),(255,0,0),20)
        
    return fix_img
```

```python
result =detect_eyes(fix)
plt.imshow(result)
```
